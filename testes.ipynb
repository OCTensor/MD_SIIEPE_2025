```python
import numpy as np

# Gera átomos aleatórios
atomos = []
for i in range(6):
    atomo = np.random.rand(3)
    atomos.append(atomo)

raio = 1.4
limite = 2 * raio

sobreposto = True
while sobreposto:
    sobreposto = False  # assume que tá ok, só muda se achar colisão
    
    for i in range(len(atomos)):
        for j in range(i + 1, len(atomos)):  # evita comparar A com A e repetições
            dist = np.linalg.norm(atomos[i] - atomos[j])
            if dist < limite:
                print(f"Átomos {i} e {j} estão sobrepostos! (dist = {dist:.2f})")
                sobreposto = True
```
