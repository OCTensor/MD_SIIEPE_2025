```python
import numpy as np

# Gera tomos aleatrios
atomos = []
for i in range(6):
    atomo = np.random.rand(3)
    atomos.append(atomo)

raio = 1.4
limite = 2 * raio

sobreposto = True
while sobreposto:
    sobreposto = False  # assume que t ok, s muda se achar coliso
    
    for i in range(len(atomos)):
        for j in range(i + 1, len(atomos)):  # evita comparar A com A e repeties
            dist = np.linalg.norm(atomos[i] - atomos[j])
            if dist < limite:
                print(f"tomos {i} e {j} esto sobrepostos! (dist = {dist:.2f})")
                sobreposto = True
```
